```python
import requests
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# ============================
# 🌡️ HEAT INDEX FUNCTION
# ============================
def calculate_heat_index(temp, humidity):
    return temp + 0.33 * humidity - 0.7

# ============================
# 🌦️ WEATHER API (FIXED)
# ============================
def get_weather(lat, lon):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current_weather=true&hourly=relative_humidity_2m,soil_moisture_1_to_3cm,precipitation"

    response = requests.get(url)
    data = response.json()

    current = data["current_weather"]
    index = -1  # latest data

    return {
        "temperature": current["temperature"],
        "wind_speed": current["windspeed"],
        "precipitation": data["hourly"]["precipitation"][index],
        "humidity": data["hourly"]["relative_humidity_2m"][index],
        "soil_moisture": data["hourly"]["soil_moisture_1_to_3cm"][index]
    }

# ============================
# 📍 PROCESS LOCATIONS
# ============================
def process_locations(locations, mode="flood"):
    data = []

    for loc in locations:
        w = get_weather(loc["lat"], loc["lon"])

        if mode == "heat":
            heat_index = calculate_heat_index(w["temperature"], w["humidity"])
            data.append({
                "location": loc["location"],
                "lat": loc["lat"],
                "lon": loc["lon"],
                "temperature": w["temperature"],
                "heat_index": heat_index
            })
        else:
            data.append({
                "location": loc["location"],
                "lat": loc["lat"],
                "lon": loc["lon"],
                **w
            })

    return pd.DataFrame(data)

# ============================
# 📍 DATASETS
# ============================
flood_locations = kerla_locations
heat_locations = bihar_locations

df_flood = process_locations(flood_locations, mode="flood")
df_heat = process_locations(heat_locations, mode="heat")

# ============================
# 🧠 LABEL CREATION (FIXED)
# ============================

# 🌊 Flood Labels (dynamic threshold)
threshold_flood = df_flood[['humidity','soil_moisture']].mean().mean()

def classify_flood(row):
    score = row['humidity']*0.5 + row['soil_moisture']*100
    return 1 if score >= threshold_flood else 0

df_flood['label'] = df_flood.apply(classify_flood, axis=1)

# 🔥 Heat Labels (dynamic)
threshold_heat = df_heat['heat_index'].mean()

def classify_heat(row):
    return 1 if row['heat_index'] >= threshold_heat else 0

df_heat['label'] = df_heat.apply(classify_heat, axis=1)

# ============================
# 🌊 FLOOD ML MODEL
# ============================
X_flood = df_flood[['wind_speed', 'precipitation', 'humidity', 'soil_moisture']]
y_flood = df_flood['label']

rf_flood = RandomForestClassifier(n_estimators=50, random_state=42)
rf_flood.fit(X_flood, y_flood)

# Safe Logistic Regression
if len(set(y_flood)) > 1:
    lr_flood = LogisticRegression()
    lr_flood.fit(X_flood, y_flood)
    use_lr_flood = True
else:
    use_lr_flood = False

rf_pred_flood = rf_flood.predict(X_flood)
rf_acc_flood = accuracy_score(y_flood, rf_pred_flood)

if use_lr_flood:
    lr_pred_flood = lr_flood.predict(X_flood)
    lr_acc_flood = accuracy_score(y_flood, lr_pred_flood)
else:
    lr_acc_flood = "Skipped (one class)"

best_flood_model = rf_flood

df_flood['ML_Prediction'] = best_flood_model.predict(X_flood)
df_flood['ML_Prediction'] = df_flood['ML_Prediction'].map({0: "Safe", 1: "Danger"})

# ============================
# 🔥 HEAT ML MODEL
# ============================
X_heat = df_heat[['temperature', 'heat_index']]
y_heat = df_heat['label']

rf_heat = RandomForestClassifier(n_estimators=50, random_state=42)
rf_heat.fit(X_heat, y_heat)

if len(set(y_heat)) > 1:
    lr_heat = LogisticRegression()
    lr_heat.fit(X_heat, y_heat)
    use_lr_heat = True
else:
    use_lr_heat = False

rf_pred_heat = rf_heat.predict(X_heat)
rf_acc_heat = accuracy_score(y_heat, rf_pred_heat)

if use_lr_heat:
    lr_pred_heat = lr_heat.predict(X_heat)
    lr_acc_heat = accuracy_score(y_heat, lr_pred_heat)
else:
    lr_acc_heat = "Skipped (one class)"

best_heat_model = rf_heat

df_heat['ML_Prediction'] = best_heat_model.predict(X_heat)
df_heat['ML_Prediction'] = df_heat['ML_Prediction'].map({0: "Safe", 1: "Heatwave"})

# ============================
# 📊 OUTPUT
# ============================
print("🌊 Flood Accuracy → RF:", rf_acc_flood, " LR:", lr_acc_flood)
display(df_flood[['location', 'precipitation', 'humidity', 'soil_moisture', 'ML_Prediction']])

print("\n🔥 Heatwave Accuracy → RF:", rf_acc_heat, " LR:", lr_acc_heat)
display(df_heat[['location', 'temperature', 'heat_index', 'ML_Prediction']])
```

```python
import pandas as pd

# ============================
# 📍 Relief Camps
# ============================
relief_df = pd.DataFrame(relief_camps)
relief_df['location'] = relief_df['name']
relief_df['zone_status'] = 'Relief Camp'

# ============================
# 🌊 FLOOD (Using ML Output)
# ============================
safe_flood_df = df_flood[df_flood['ML_Prediction'] == 'Safe'].reset_index(drop=True)
danger_flood_df = df_flood[df_flood['ML_Prediction'] == 'Danger'].copy()

# ============================
# 🌊 FLOOD SEVERITY
# ============================
def get_score(value, thresholds):
    for i, threshold in enumerate(thresholds):
        if value <= threshold:
            return i
    return len(thresholds)

def calculate_flood_severity(row):
    wind_score = get_score(row['wind_speed'], [7, 10, 20])
    precip_score = get_score(row['precipitation'], [3, 5, 7])
    humidity_score = get_score(row['humidity'], [84, 90, 95])
    soil_score = get_score(row['soil_moisture'], [0.35, 0.5, 0.6])
    return wind_score + precip_score + humidity_score + soil_score

def flood_severity_label(score):
    if score <= 2:
        return 'Low'
    elif score <= 5:
        return 'Medium'
    else:
        return 'High'

danger_flood_df['severity_score'] = danger_flood_df.apply(calculate_flood_severity, axis=1)
danger_flood_df['severity_level'] = danger_flood_df['severity_score'].apply(flood_severity_label)

# ============================
# 🔥 HEATWAVE (Using ML Output)
# ============================
safe_heat_df = df_heat[df_heat['ML_Prediction'] == 'Safe'].reset_index(drop=True)
danger_heat_df = df_heat[df_heat['ML_Prediction'] == 'Heatwave'].copy()

# ============================
# 🔥 HEATWAVE SEVERITY
# ============================
def calculate_heat_severity(row):
    # Only based on heat index
    if row['temperature'] < 35:
        return 1
    elif row['temperature'] < 45:
        return 2
    else:
        return 3

def heat_severity_label(score):
    if score == 1:
        return 'Low'
    elif score == 2:
        return 'Medium'
    else:
        return 'High'

danger_heat_df['severity_score'] = danger_heat_df.apply(calculate_heat_severity, axis=1)
danger_heat_df['severity_level'] = danger_heat_df['severity_score'].apply(heat_severity_label)

# ============================
# 📊 OUTPUT
# ============================

print("🌊 Flood Danger Zones with Severity")
display(danger_flood_df[['location','wind_speed','precipitation', 'humidity', 'soil_moisture', 'severity_level']])

print("\n🔥 Heatwave Zones with Severity")
display(danger_heat_df[['location', 'temperature', 'heat_index', 'severity_level']])
```

```python
import math

def shift_location(lat, lon, shift_meters_north=1000, shift_meters_east=1000):
    delta_lat = shift_meters_north / 111000
    delta_lon = shift_meters_east / (111000 * math.cos(math.radians(lat)))
    return lat + delta_lat, lon + delta_lon


relief_camps = []
for loc in kerla_locations:
    shifted_lat, shifted_lon = shift_location(loc['lat'], loc['lon'], 1000, 1000)
    relief_camps.append({
        "name": f"{loc['location']} Relief Camp",
        "lat": shifted_lat,
        "lon": shifted_lon
    })

for camp in relief_camps:
    print(camp)
```

```python
import os
import folium
import pandas as pd
from geopy.distance import geodesic
from deep_translator import GoogleTranslator
import openrouteservice
from openrouteservice import convert
import json
from folium.plugins import MiniMap, MeasureControl

client = openrouteservice.Client(key=os.environ.get('ORS_API_KEY', 'YOUR_ORS_API_KEY'))

# ============================
# 🌐 LANGUAGE SUPPORT
# ============================
LANGUAGES = {
    "English": "en", "Hindi": "hi", "Tamil": "ta", "Telugu": "te", "Kannada": "kn",
    "Malayalam": "ml", "Bengali": "bn", "Marathi": "mr", "Gujarati": "gu",
    "Punjabi": "pa", "Urdu": "ur", "Assamese": "as", "Odia": "or"
}

# ============================
# 🔹 DATA (ONLY FLOOD)
# ============================
safe_df = df_flood[df_flood['ML_Prediction'] == 'Safe'].reset_index(drop=True)
danger_df = danger_flood_df.copy()

relief_df = pd.DataFrame(relief_camps)
relief_df['location'] = relief_df['name']

# ============================
# MAP CENTER
# ============================
avg_lat, avg_lon = df_flood['lat'].mean(), df_flood['lon'].mean()
m = folium.Map(location=[avg_lat, avg_lon], zoom_start=8)

# ============================
# LANGUAGE DROPDOWN
# ============================
dropdown_html = f"""
<div style="position: fixed; top: 10px; left: 50px; z-index: 1000;">
  <label><b>Select Language:</b></label>
  <select onchange="changeLanguage(this.value)">
    {''.join([f'<option value="{code}">{lang}</option>' for lang, code in LANGUAGES.items()])}
  </select>
</div>
<script>
function changeLanguage(lang) {{
  const elements = document.querySelectorAll("[data-translate]");
  elements.forEach(el => {{
    const translations = JSON.parse(el.getAttribute("data-translate"));
    el.innerHTML = translations[lang] || translations['en'];
  }});
}}
</script>
"""
m.get_root().html.add_child(folium.Element(dropdown_html))

# ============================
# TRANSLATION
# ============================
def make_translations(text_dict):
    translations = {}
    for _, code in LANGUAGES.items():
        try:
            translations[code] = GoogleTranslator(source='en', target=code).translate(text_dict['en'])
        except:
            translations[code] = text_dict['en']
    return translations

def translated_popup(text_dict):
    translations = make_translations(text_dict)
    escaped = json.dumps(translations).replace('"', '&quot;')
    html = f'<div data-translate="{escaped}">{translations["en"]}</div>'
    return folium.Popup(html, max_width=300)

# ============================
# SAFE ZONES
# ============================
for _, row in safe_df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=50,
        color='green',
        fill=True,
        fill_opacity=0.4,
        popup=translated_popup({"en": f"✅ Safe Zone: {row['location']}"})
    ).add_to(m)

# ============================
# RELIEF CAMPS
# ============================
for _, row in relief_df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        icon=folium.Icon(color='blue'),
        popup=translated_popup({"en": f"🏫 Relief Camp: {row['location']}"})
    ).add_to(m)

# ============================
# DANGER ZONES
# ============================
colors = {'Low': 'yellow', 'Medium': 'orange', 'High': 'red'}

for _, row in danger_df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=50,
        color=colors[row['severity_level']],
        fill=True,
        fill_opacity=0.3,
        popup=translated_popup({
            "en": f"🚨 Flood Zone: {row['location']}<br>Severity: {row['severity_level']}"
        })
    ).add_to(m)

# ============================
# ROUTING FUNCTION (UNCHANGED)
# ============================
def get_route_info(start_lat, start_lon, end_lat, end_lon):
    coords = [(start_lon, start_lat), (end_lon, end_lat)]
    routes = client.directions(coords)
    geometry = routes['routes'][0]['geometry']
    duration = routes['routes'][0]['summary']['duration']
    distance = routes['routes'][0]['summary']['distance']
    decoded = convert.decode_polyline(geometry)
    return [(lat, lon) for lon, lat in decoded['coordinates']], duration, distance


# ============================
# 🚨 DANGER → RELIEF (BLUE)
# (OLD LOGIC STYLE ✅)
# ============================
for _, danger in danger_df.iterrows():
    danger_loc = (danger['lat'], danger['lon'])

    # ⚠️ IMPORTANT: create temp distance column (NOT overwrite original permanently)
    relief_df['temp_dist'] = relief_df.apply(
        lambda x: geodesic(danger_loc, (x['lat'], x['lon'])).km, axis=1
    )

    nearest_relief = relief_df.loc[relief_df['temp_dist'].idxmin()]

    try:
        route_coords, duration, dist = get_route_info(
            danger['lat'], danger['lon'],
            nearest_relief['lat'], nearest_relief['lon']
        )

        text = {
            "en": f"<b>Route Info</b><br>From: {danger['location']}<br>To: {nearest_relief['location']}<br>Distance: {round(dist/1000,2)} km<br>Duration: {round(duration/60,1)} min"
        }

        folium.PolyLine(
            route_coords,
            color='blue',
            weight=4,
            popup=translated_popup(text)
        ).add_to(m)

    except Exception as e:
        print("Routing error:", e)


# ============================
# 🟣 SAFE → NEAREST DANGER
# (OLD LOGIC STYLE ✅)
# ============================
for _, safe in safe_df.iterrows():
    safe_loc = (safe['lat'], safe['lon'])

    # ⚠️ IMPORTANT: use temp column (NOT overwrite main df)
    temp_df = danger_df.copy()

    temp_df['temp_dist'] = temp_df.apply(
        lambda x: geodesic(safe_loc, (x['lat'], x['lon'])).km, axis=1
    )

    nearest_dangers = temp_df.nsmallest(3, 'temp_dist')

    for _, danger in nearest_dangers.iterrows():
        try:
            route_coords, duration, dist = get_route_info(
                safe['lat'], safe['lon'],
                danger['lat'], danger['lon']
            )

            text = {
                "en": f"<b>Route Info</b><br>From: {safe['location']}<br>To: {danger['location']}<br>Distance: {round(dist/1000,2)} km<br>Duration: {round(duration/60,1)} min"
            }

            folium.PolyLine(
                route_coords,
                color='purple',
                weight=4,
                popup=translated_popup(text)
            ).add_to(m)

        except Exception as e:
            print("Routing error:", e)

# ============================
# 🟣 SAFE → NEAREST DANGER
# ============================
for _, safe in safe_df.iterrows():
    safe_loc = (safe['lat'], safe['lon'])

    danger_df['distance'] = danger_df.apply(
        lambda x: geodesic(safe_loc, (x['lat'], x['lon'])).km, axis=1
    )
    nearest_dangers = danger_df.nsmallest(3, 'distance')

    for _, danger in nearest_dangers.iterrows():
        try:
            route_coords, duration, dist = get_route_info(
                safe['lat'], safe['lon'],
                danger['lat'], danger['lon']
            )

            text = {"en": f"<b>Nearby Risk</b><br>From: {safe['location']}<br>To: {danger['location']}<br>Distance: {round(dist/1000,2)} km"}

            folium.PolyLine(route_coords, color='purple', weight=3,
                            popup=translated_popup(text)).add_to(m)

        except:
            pass

# ============================
# CONTROLS
# ============================
m.add_child(MiniMap())
m.add_child(MeasureControl())

# ============================
# 🔥 HEATWAVE MAP
# ============================

# Data
safe_heat_df = df_heat[df_heat['ML_Prediction'] == 'Safe'].reset_index(drop=True)
danger_heat_df = danger_heat_df.copy()

# Map center (heat only)
avg_lat_h, avg_lon_h = df_heat['lat'].mean(), df_heat['lon'].mean()

map_heat = folium.Map(
    location=[avg_lat_h, avg_lon_h],
    zoom_start=6,
    tiles='CartoDB positron'  # avoids 403 error
)

# ============================
# LANGUAGE DROPDOWN (SAME)
# ============================
dropdown_html_heat = f"""
<div style="position: fixed; top: 10px; left: 50px; z-index: 1000;">
  <label><b>Select Language:</b></label>
  <select onchange="changeLanguage(this.value)">
    {''.join([f'<option value="{code}">{lang}</option>' for lang, code in LANGUAGES.items()])}
  </select>
</div>
<script>
function changeLanguage(lang) {{
  const elements = document.querySelectorAll("[data-translate]");
  elements.forEach(el => {{
    const translations = JSON.parse(el.getAttribute("data-translate"));
    el.innerHTML = translations[lang] || translations['en'];
  }});
}}
</script>
"""
map_heat.get_root().html.add_child(folium.Element(dropdown_html_heat))

# ============================
# SAFE HEAT ZONES
# ============================
for _, row in safe_heat_df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=40,
        color='green',
        fill=True,
        fill_opacity=0.3,
        popup=translated_popup({"en": f"✅ Safe Area: {row['location']}"})
    ).add_to(map_heat)

# ============================
# HEATWAVE ZONES (SEVERITY)
# ============================
colors_heat = {'Low': 'yellow', 'Medium': 'orange', 'High': 'red'}

for _, row in danger_heat_df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=50,
        color=colors_heat[row['severity_level']],
        fill=True,
        fill_opacity=0.4,
        popup=translated_popup({
            "en": f"🔥 Heatwave: {row['location']}<br>Severity: {row['severity_level']}<br>Temp: {row['temperature']}°C"
        })
    ).add_to(map_heat)

# ============================
# CONTROLS
# ============================
map_heat.add_child(MiniMap())
map_heat.add_child(MeasureControl())

# ============================
# OUTPUT
# ============================

print("🌊 Flood Map")
display(m)

print("🔥 Heatwave Map")
display(map_heat)
```
